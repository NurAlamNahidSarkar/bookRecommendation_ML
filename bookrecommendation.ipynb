# **Popular Books Using Normal Average**

```python
book_num_ratings = book_df.groupby('Book-Title').count()['Book-Rating'].reset_index()
book_num_ratings.rename(columns = {'Book-Rating' : 'Number of Ratings'}, inplace = True)
book_num_ratings
```

```python
book_avg_ratings = book_df.groupby('Book-Title')['Book-Rating'].mean().reset_index()
book_avg_ratings.rename(columns = {'Book-Rating' : 'Average of Ratings'}, inplace = True)
book_avg_ratings
```

```python
popular_books = book_num_ratings.merge(book_avg_ratings, on = 'Book-Title')
popular_books
```

```python
popular_books = popular_books[popular_books['Number of Ratings'] >= 200].sort_values('Average of Ratings', ascending = False).reset_index()
```

# **Bayesian Average Formula**

```python
# C = The average rating across ALL books in the dataset
# (Note: Use your main ratings DataFrame here, not the grouped one)
C = book_df['Book-Rating'].mean()

# m = The minimum number of ratings to be considered.
# We'll use 200 as you suggested. This is our "gatekeeper" threshold.
m = 200

print(f"Global Average Rating (C): {C}")
print(f"Minimum Ratings (m): {m}")

#Get 'v' (vote count) and 'R' (average) for each book ---

# Group by ISBN to get the count (v) and mean (R) for each book
book_stats = book_df.groupby('ISBN')['Book-Rating'].agg(['count', 'mean'])
book_stats.rename(columns={'count': 'v', 'mean': 'R'}, inplace=True)

# Apply the Bayesian Average Formula ---

def weighted_rating(row, m, C):
    v = row['v'] # Number of ratings for this book
    R = row['R'] # Average rating for this book
    
    return (v / (v + m)) * R + (m / (v + m)) * C

book_stats['Weighted Rating'] = book_stats.apply(weighted_rating, args=(m, C), axis=1)
top_books = book_stats.sort_values('Weighted Rating', ascending=False)

book_info = book_df.drop_duplicates(subset='ISBN')
top_books = top_books.merge(book_info, on='ISBN')

print("\n--- Top Books by Weighted Rating ---")
print(top_books[['Book-Title', 'v', 'R', 'Weighted Rating']].head(10))
```

```python
# Sort by the new 'Weighted Rating' column
top_books = book_stats.sort_values('Weighted Rating', ascending=False)
```
