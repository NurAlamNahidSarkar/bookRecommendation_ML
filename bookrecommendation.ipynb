```python
books = pd.read_csv('/kaggle/input/book-recommendation-dataset/Books.csv')
ratings = pd.read_csv('/kaggle/input/book-recommendation-dataset/Ratings.csv')
users = pd.read_csv('/kaggle/input/book-recommendation-dataset/Users.csv')
```

```python
ratings_users = ratings.merge(users, on = 'User-ID')
```

```python
ratings_users.shape
```

```python
book_df = books.merge(ratings_users, on = 'ISBN')
```

```python
book_df.shape
```

```python
book_df.head()
```

```python
book_df.isna().mean()*100
```

```python
null_counts = book_df.isna().sum()
cols_with_nulls = null_counts[null_counts > 0].index.tolist()
cols_with_nulls = [col for col in cols_with_nulls if col != 'Age']

print(f"Columns with nulls: {cols_with_nulls}")
```

```python
book_df = book_df.dropna(subset = cols_with_nulls)
```

```python
book_df.isna().mean()*100
```

```python
age_mean = book_df['Age'].mean()
```

```python
book_df['Age'] = book_df['Age'].fillna(age_mean)
```
